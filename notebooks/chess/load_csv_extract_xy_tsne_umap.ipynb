```python
from openTSNE import TSNE
from openTSNE.callbacks import ErrorApproximations
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
file = 'openings30.csv'
```

```python
df = pd.read_csv(file, header=0)
```

```python
fields = ['a8','b8','c8','d8','e8','f8','g8','h8','a7','b7','c7','d7','e7','f7','g7','h7','a6','b6','c6','d6','e6','f6','g6','h6','a5','b5','c5','d5','e5','f5','g5','h5','a4','b4','c4','d4','e4','f4','g4','h4','a3','b3','c3','d3','e3','f3','g3','h3','a2','b2','c2','d2','e2','f2','g2','h2','a1','b1','c1','d1','e1','f1','g1','h1']
print(len(df[fields]))
df[fields].head()
```

```python
df = df[fields].fillna('')
df.head()
```

```python
data = df[fields].to_numpy()
print(data)
```

```python
def one_hot(array):
    unique, inverse = np.unique(array, return_inverse=True)
    onehot = np.eye(unique.shape[0])[inverse]
    return onehot

oh_data = one_hot(data)
data = oh_data.reshape(data.shape[0], -1)
# print(data)
print(data.shape)
```

```python
seed = 42
tsne = TSNE(
    perplexity=200,
    n_jobs=6,
    metric='euclidean',
    random_state=seed,
    verbose=True
)
```
