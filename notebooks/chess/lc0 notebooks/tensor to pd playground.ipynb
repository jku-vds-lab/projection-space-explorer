```python
import torch as th
import numpy as np
import pandas as pd

FILE_NAME = 'tensors.csv'
```

## create tensor

```python
x = th.rand(10,128,8,8)
```

## saving tensor to pd

first creating empty frame, then appending

```python
df = pd.DataFrame()
df.append(pd.DataFrame(x.reshape(-1,128*8*8).numpy()))
```

or creating directly from data

```python
df = pd.DataFrame(x.reshape(-1,128*8*8).numpy())
df.head()
```

```python
df.info()
```

## iteratively add to tensor

```python
x2 = th.rand(10,128,8,8)
df = df.append(pd.DataFrame(x2.reshape(-1,128*8*8).numpy()))
df.info()
```

## storing as csv

```python
df.to_csv(FILE_NAME)
```

## loading

```python
df = pd.read_csv(FILE_NAME)
df.head()
```

```python
df.info()
```

```python
loaded_x = th.tensor(df.values[:,1:])
print(loaded_x)
print(loaded_x.shape)
```

## umap on this

```python
import umap.umap_ as umap
```

```python
reducer = umap.UMAP()
```

```python
final_data = loaded_x
umap_embedding = reducer.fit_transform(final_data)
```

```python
print(umap_embedding)
```

```python
 from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax.set_aspect(1)
ax.scatter(umap_embedding[:,0], umap_embedding[:,1], alpha=0.1, color='red')
# for game in umap_embedding_split[:100]:
#     ax.plot(game[:,0], game[:,1], '-r', alpha=0.1, color='red')
# for game in umap_embedding_split[100:]:
#     ax.plot(game[:,0], game[:,1], '-r', alpha=0.1, color='blue')
```
