This was used for the multi-level embedding paper re-submission.
We read the original JSON file that contains the premade groups, do some magic to actually be able to access the data within the JSON, copy the already existing algo feature, map it to opening names, do some more magic to put it back into the JSON, and store at in the original JSON format.

```python
import numpy as np
import pandas as pd
file = 'chess40k_groups.json'
```

## pd.DataFrame from JSON

```python
df_json = pd.read_json(file)
df_dict = df_json.to_dict()

header = df_dict['samples'][0]['columns']
data = df_dict['samples'][0]['data']

df = pd.DataFrame(data=data, columns=header)
df.head()
```

```python
# print(data[:50])
print(header)
```

## Duplicate algo into new opening column and replace by name

```python
df['Opening'] = df['algo']
df['Opening'].replace({0: 'Zukertort', 1: 'Queen\'s Pawn', 2: 'English'}, inplace=True)
df.head()
```

## overwrite json and save

```python
import json
```

```python
data = df.values.tolist()
header = list(df.columns.values)
df_dict['samples'][0]['columns'] = header
df_dict['samples'][0]['data'] = data
df_dict
```

```python
df_json
```

```python
df_json = pd.DataFrame.from_dict(df_dict)
df_json
```

```python
df_json.to_json('updated_openings.json')
```
